```r
library(googlesheets4)
library(ggplot2)
```

```r
gs4_deauth()
```

```r
myDF <- read_sheet('https://docs.google.com/spreadsheets/d/1y_g37nbDgHC-zm1y_WGfThBt5wipMxnFSDcSZT6AQq4/edit?usp=sharing')
```

```r
str(myDF)
```

```r
summary(myDF)
```

```r
unique(myDF$Year)
```

```r
sort(table(myDF$Year))
```

```r
year4data <- myDF[myDF$Year == "Year 4",]
year4orders <- table(year4data$Order)

ggplot(data = data.frame(Order = names(year4orders), Freq = as.numeric(year4orders)), 
       aes(x = Order, y = Freq)) +
  geom_bar(stat = "identity", fill = "lightblue") +
  labs(title = "Food Orders from Year 4 Students",
       x = "Food Item",
       y = "Number of Orders") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
